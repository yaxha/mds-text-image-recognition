```python
#La función asume que se le está enviando una imagen en RGB
def problema1(imagen, color):
    ancho = imagen.shape[0]
    alto = imagen.shape[1]
    
    salida = np.zeros((ancho, alto, 3))
    
    activo_azul = 0
    activo_verde = 0
    activo_rojo = 0
    
    if(color == 1):
        activo_azul = 1
    elif(color == 2):
        activo_verde = 1
    elif(color == 3):
        activo_rojo = 1
    elif(color == 10):
        activo_verde = 1
        activo_rojo = 1
    elif(color == 20):
        activo_azul = 1
        activo_verde = 1
    elif(color == 30):
        activo_azul = 1
        activo_rojo = 1
    elif(color == 40): #Prueba para determinar si devuelve la imagen original
        activo_azul = 1
        activo_rojo = 1
        activo_verde = 1
    
    salida[:,:,0] = activo_azul * imagen[:,:,2]
    salida[:,:,1] = activo_verde * imagen[:,:,1]
    salida[:,:,2] = activo_rojo * imagen[:,:,0]
    
    cv2.imwrite('problema1/problema1.jpg', salida)
    
    imagen_salida = cv2.imread('problema1/problema1.jpg')
    imagen_salida = cv2.cvtColor(imagen_salida, cv2.COLOR_BGR2RGB)
    plt.imshow(imagen_salida)
    plt.show()

    #return salida
```

## Problema 4

> Nota: entiendo que no podemos usar opencv para calcular el promedio aritmético, pero sí para hallar las distribuciones

```python
def problema4(carpeta, imagen):
    img = cv2.imread(carpeta + imagen)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = np.average(img.astype(np.int), axis=2)
    cv2.imwrite(carpeta + 'gris.jpg', gray)
    
    imagen_gris = cv2.imread(carpeta + 'gris.jpg')
    plt.imshow(imagen_gris)
    plt.show()
    
    chans = cv2.split(img)
    colors = ("b", "g", "r")
    plt.figure()
    plt.title("Histograma de colores")
    plt.xlabel("Intensidad del píxel")
    plt.ylabel("Cantidad de píxeles")
    features = []
    # loop over the image channels
    for (chan, color) in zip(chans, colors):
        hist = cv2.calcHist([chan], [0], None, [256], [0, 256])
        plt.plot(hist, color = color)
        plt.xlim([0, 256])
        plt.show()
    hist = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])
    plt.plot(hist, color = "gray")
    plt.xlim([0, 256])
    plt.show()
    
```

## Problema 5
Investigue en qué consiste el enfoque de escala de grises ponderado, luego de esto implemente una función que dada una imagen,  realice una escala de grises ponderada (notar que no existe una solución única).

### Enfoques comunes
A la hora de aplicar escalas de grises a una imágen, se suelen aplicar tres formulas, estas son:
* Lightness: Método que calcula la media de los colores extremos. 
    (max(R,G,B) + min(R,G,B)) / 2
* Average: Media de cada unos de los colores que componen la imagen.
    (R, G, B) / 3
* Luminosity: Media ponderada que considera la forma en que los humanos percibimos los colores.
    0.21 R + 0.71 G + 0.07 B
    
La visión humana es más sensible al verde, por lo que tiene el mayor valor de coeficiente y es menos sensible al azul, por lo que tiene el menor coeficiente.

```python
def problema5(carpeta, imagen):
    img = cv2.imread(carpeta + imagen)
    ancho, alto, _ = img.shape
    salida = np.zeros((ancho, alto, 1))
    
    salida[:,:,0] = 0.0722 * img[:,:,0] + 0.7152 * img[:,:,1] + 0.2126 * img[:,:,2]
    cv2.imwrite(carpeta + 'resultado.jpg', salida)
    
    imagen_salida = cv2.imread(carpeta + 'resultado.jpg')
    plt.imshow(imagen_salida)
    plt.show()
```

## Problema 6
Investigue brevemente en qué consiste el espacio de color HSV y cómo se mapean colores a dicho espacio, para entregar este ejercicio puede hacer con un archivo PDF.

### RGB
OpenCV utiliza RGB (Red Green Blue) como espacio de colores predeterminado, aunque en realidad almacena las imagenes en BGR (intercambiando los colores azul y rojo). Esta forma de almacenar las imagenes es sencilla ya que crea 3 capas de cada imagen y a cada píxel le asigna una intensidad de 0 a 255 para cada color.

### HSV
El espacio de color HSV proviene de las siglas Hue (matiz), Saturation, Valor.
* Hue – Este canal codifica la tonalidad del color: rojo, azul, amarillo, verde, etc. En OpenCV el valor del Hue está entre 0-179. Es muy útil en tareas de procesamiento de imágenes que necesitan segmentar objetos en función de su color.
* Saturation – Este canal codifica la intensidad o pureza del color: cuanta menos saturación más grisáceo será el color. En OpenCV el valor de Saturation oscila entre 0-255.
* Value – Este canal codifica la luminosidad: cuanto menor sea, más negro/oscuro será el color.

El canal Hue es en realidad un escalar que va de 0 a 360 _grados_, pero OpenCV lo reduce de 0 a 179. La saturación (S) varía desde el centro hacia afuera y el valor (V) desde lo más oscuro, la base, hasta lo más claro, el techo.
